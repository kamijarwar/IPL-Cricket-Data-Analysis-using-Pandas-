```python
team1 = df["Team1"].value_counts()
team2 = df["Team2"].value_counts()

matches_played = team1.add(team2, fill_value=0)
```

```python
matches_won = df["WinningTeam"].value_counts()
```

```python
winning_percentage = ((matches_won / matches_played) * 100).round(2)

winning_percentage = winning_percentage.sort_values(ascending=False)

print(winning_percentage)
```

```python
matches_per_season = df["Season"].value_counts().sort_index()

plt.figure(figsize=(12,6))

plt.plot(matches_per_season.index,
         matches_per_season.values,
         marker='o')

plt.title("IPL Matches Played Per Season")

plt.xlabel("Season")

plt.ylabel("Number of Matches")

plt.grid(True)

plt.show()
```

```python
top_teams = df["WinningTeam"].value_counts().head(10)

plt.figure(figsize=(12,6))

plt.bar(top_teams.index,
        top_teams.values)

plt.title("Top 10 Most Successful IPL Teams")

plt.xlabel("Teams")

plt.ylabel("Total Wins")

plt.xticks(rotation=45)

plt.show()
```
